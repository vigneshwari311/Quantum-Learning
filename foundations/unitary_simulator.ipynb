```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator, Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(precision=3,suppress=True)
```

```python
def show_unitary(circuit):
    backend = AerSimulator(method='unitary')
    
    qc_unitary = circuit.copy()
    qc_unitary.save_unitary()
    
    tc = transpile(qc_unitary,backend=backend)
    result = backend.run(tc).result()
    unitary_matrix = result.get_unitary()
    print(f"Unitary matrix: \n{unitary_matrix}")
    return unitary_matrix
```

```python
def calculate_unitary(circuit,unit):
    shots = 1024
    num_qubits = circuit.num_qubits
    dim = 2 ** num_qubits
    initial_vector = np.zeros(dim)
    initial_vector[0] = 1
    calc_counts = {}
    
    final_statevector = np.dot(unit,initial_vector)
    
    for i in range(dim):
        prob = abs(final_statevector[i])**2
        binary_label = format(i,f'0{num_qubits}b')
        calc_counts[binary_label] = int(prob*shots)
        
    print("\nCalculated counts: ",calc_counts)
    qc_measure = circuit.copy()
    qc_measure.measure_all()
    
    backend = AerSimulator()
    tc = transpile(qc_measure,backend=backend)
    result = backend.run(tc,shots=shots).result()
    counts = result.get_counts()
    
    labels = list(counts.keys())
    values = list(counts.values())
    
    fig = plot_histogram(counts) 
    display(fig)      
    return counts
        
    
```

```python
def main():
    print("Enter the number of circuits to explore")
    user_input = int(input("0 - Exit ; 1 - One qubit superposition ; 2 - Two qubit superposition ; 3 - Two qubit entanglement"))
    if user_input == 0:
        pass
    circuit = QuantumCircuit(user_input,user_input)
    if user_input == 1:
        circuit.h(0)
    elif user_input == 2:
        circuit.h([0,1])
    else:
        circuit.h(0)
        circuit.cx(0,1)
        
    print("\nCircuit Diagram:")
    print(circuit.draw())
    
    unitary_matrix = show_unitary(circuit)
    calculate_unitary(circuit,unitary_matrix)
```

```python
if __name__ == '__main__':
    main()
    
```
